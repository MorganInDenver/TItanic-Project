# Titanic Data Visualization

In this notebook, I have great fun with Seaborn exploring the titanic dataset. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
df = pd.read_csv('./titanic_for_viz.csv', index_col=0)
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), annot=True,cmap="YlGnBu")
plt.title('Heatmap of Titanic Dataset Variables', fontsize=15)
```

```python
df['Age'].hist()
plt.title('Age of Titanic Passengers', fontsize=15)
plt.xlabel('Age')
```

```python
plt.scatter(df['Age'], df['Fare'])
plt.title('No Clear Relationship Between Fare and Age', fontsize=15)
plt.xlabel('Age')
plt.ylabel("Fare")
```

```python
sns.barplot(x="Embarked", y="Survived", data=df, ci=None)
plt.title('Survival Rate Greatest from Cherbourg', fontsize=15)
plt.xlabel('Port')
plt.ylabel("Surival Rate")
```

```python
sns.barplot(x="CabinType", y="Survived", data=df, ci=None)
plt.title('No Clear Relationship between Survival and CabinType', fontsize=15)
plt.xlabel('CabinType')
plt.ylabel("Surival Rate")
```

```python
sns.barplot(x="Pclass", y="Survived", data=df, ci=None)
plt.title('Survival Rate by Passenger Class', fontsize=15)
plt.xlabel('Pclass')
```

#### Survival of Female Passengers

These four charts clearly show that survival rates for female passengers was larger than males across Passenger Class, whether traveling with family, and across age. 

```python
fig, axs = plt.subplots(ncols=4)
fig.set_size_inches(13, 7)
sns.barplot(x="Sex", y="Survived", data=df, ci=None, ax=axs[0])
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=df, ci=None, ax=axs[1])
sns.barplot(x="Sex", y="Survived", hue="Family", data=df, ci=None, ax=axs[2])
sns.barplot(x="Sex", y="Survived", hue="age_bins", data=df, ci=None, ax=axs[3])
```
